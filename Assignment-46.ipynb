### Q1. Explain the concept of R-squared in linear regression models. How is it calculated, and what does it represent?

R-squared is a statistical measure that represents the proportion of the variance in the dependent variable that is explained by the independent variable(slinear regression model. It ranges from 0 to 1, with 1 indicating a perfect fit of the model to the data. It is calculated by dividing by the total variance of the dependent variable.

Adjusted R-squared is more appropriate when comparing multiple regression models with different numbers It adjusts for the number of variables in the model, penalizing models with too many variables that do not contribute significantly to the model's overall fit.

### Q5. Discuss the advantages and disadvantages of using RMSE, MSE, and MAE as evaluation metrics in regression analysis.

RMSE and MSE take into account the magnitude of errors, giving more weight to larger errors. This means that these metrics can be more sensitive which are data points that are significantly different from other data points. On the other hand, MAE treats all errors equally and is more robust One disadvantage of RMSE, MSE, and MAE is that they do not provide information about the direction of errors, i.e., whether the model underestimating the target variable. Another disadvantage is that they do not take into account the relative importance of different types of errors.
